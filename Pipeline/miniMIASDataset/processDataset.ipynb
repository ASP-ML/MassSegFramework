### Rename Files

```python
import os
user_path = 'C:/Users/DELL/'
directory = user_path + 'MassSegFramework/Pipeline/miniMIAS/OriginalTestImages'
files = os.listdir(directory)

for file in files:
    if file.startswith("AnyConv.com__"):
        new_name = file.replace("AnyConv.com__", "")

        original_path = os.path.join(directory, file)
        new_path = os.path.join(directory, new_name)
        os.rename(original_path, new_path)
```
